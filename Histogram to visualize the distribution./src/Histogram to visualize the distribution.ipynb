# Task-01

Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable, such as the distribution of ages or genders in a population.

Sample Dataset :-
https://data.worldbank.org/indicator/SP.POP.TOTL

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
data=pd.read_csv("Population.csv")
df=pd.DataFrame(data)
df
```

```python
df.dropna(inplace=True)
df.describe()
```

```python
df.info()
```

```python
df= df.dropna(axis=0, inplace=False).reset_index(drop=True)
df
```

```python
population_2020 = df[['Country Name', '2020']]
population_2020 = population_2020.dropna(subset=['2020'])
population_2020 = population_2020.sort_values(by='2020', ascending=False)
plt.figure(figsize=(12, 8))
colors = ['red', 'blue', 'black', 'skyblue', 'yellow', 'pink', 'green', 'blue', 'pink', 'black']
plt.bar(population_2020['Country Name'][:10], population_2020['2020'][:10], color=colors)
plt.plot(population_2020['Country Name'][:10], population_2020['2020'][:10], marker='o', linestyle='--', color='grey')
plt.xlabel('Region')
plt.ylabel('Population in 2020')
plt.title('Top 10 Region by Population in 2020')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
df = df[['Country Name', '2021']][:10]
df = df.sort_values(by='2021', ascending=False)
plt.figure(figsize=(12, 8))
sns.barplot(x='2021', y='Country Name', data=df, palette='viridis')
plt.title(f'Top 10 Countries Based on Population in {2021}')
plt.xlabel('Population')
plt.ylabel('Country')
plt.show()
```

```python
data=pd.read_csv("Population.csv")
df=pd.DataFrame(data)
df.dropna(inplace=True)
df= df.dropna(axis=0, inplace=False).reset_index(drop=True)
aruba_data = df[df['Country Name'] == 'Aruba']
years = df.columns[-10:]
population = aruba_data.iloc[:, -10:]
population = pd.to_numeric(population.values.flatten())
plt.figure(figsize=(10, 6))
plt.bar(years, population, color='skyblue')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population of Aruba (2013-2022)')
plt.show()
```

```python
plt.plot(years, population,color="blue")
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population of Aruba (2013-2022)')
plt.show()
```

```python

sns.histplot(population, bins=10, color='skyblue', edgecolor='black')
plt.show()
```

```python
top_10_countries = df.nlargest(10, '2021')
plt.figure(figsize=(10, 6))
sns.barplot(data=top_10_countries, x='Country Code', y='2021', color='skyblue', edgecolor='black')
```

```python
least_10_countries = df.nsmallest(10, '2021')
plt.figure(figsize=(10, 6))
sns.barplot(data=least_10_countries, x='Country Code', y='2021', color='pink', edgecolor='black')
plt.title('Least 10 Countries in 2021')
```
