```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
data=pd.read_csv("/home/mustafa/internship/PRODIGY_DS_INTERN/PRODIGY_DS_Task-02/Datasets/train.csv")
df=pd.DataFrame(data)
df.head(10)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.dropna()
```

```python
df.isnull()
```

```python
df.hist(figsize=(10, 8))
```

```python
sns.pairplot(df)
```

```python
correlation = df.corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm')
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Sex', data=df)
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Age', data=df)
```

```python
# Remove outliers using IQR 
Q1 = df['Age'].quantile(0.25)
Q3 = df['Fare'].quantile(0.75)
IQR = Q3 - Q1
df = df[(df['Age'] >= Q1 - 1.5 * IQR) & (df['Fare'] <= Q3 + 1.5 * IQR)]
df
```

```python
df.to_csv('cleaned_data.csv', index=False)
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Survived', data=df, palette='viridis')
plt.xlabel('Survived')
```

```python
sns.countplot(x='Pclass', data=df, palette='Set1')
plt.title('Passenger Class Distribution')
```

```python
plt.figure(figsize=(12, 10))
sns.pairplot(df, hue='Survived', palette='Set1')
plt.suptitle('Pairplot of Numerical Variables', y=1.02)
```

```python
sns.countplot(x='Pclass', hue='Survived', data=df, palette='viridis')
plt.title('Survival Count by Passenger Class')
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Sex', hue='Survived', data=df, palette='muted')
plt.title('Survival Count by Gender')
```

```python
sns.histplot(df, x='Age', hue='Survived', bins=20, kde=True, palette='viridis')
plt.title('Age Distribution by Survival')
```

```python
sns.countplot(x='Embarked', hue='Survived', data=df, palette='dark')
plt.title('Survival Count by Embarked Port')
```
